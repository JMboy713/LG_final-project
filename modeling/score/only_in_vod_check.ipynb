```python
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
from matplotlib import font_manager, rc
import platform

if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname = 'c:/Windows/Fonts/malgun.ttf').get_name()
    rc('font', family=font_name)
```

```python
con_08 = pd.read_csv('../data/데이터스쿨3차_2308월/데이터스쿨_3차_콘텐츠_2308.csv', encoding = 'cp949', sep = '\t')
con_09 = pd.read_csv('../data/데이터스쿨3차_2309월/데이터스쿨_3차_콘텐츠_2309.csv', encoding = 'cp949', sep = '\t')
vod_08 = pd.read_csv('../data/데이터스쿨3차_2308월/데이터스쿨_3차_VOD_2308.csv', encoding = 'cp949', sep = '\t')
vod_09 = pd.read_csv('../data/데이터스쿨3차_2309월/데이터스쿨_3차_VOD_2309.csv', encoding = 'cp949', sep = '\t')
```

```python
# 8,9월 데이터 합치기
con_89 = pd.concat([con_08, con_09], ignore_index=True)
con_89.head(3)
```

```python
# 8,9월 데이터 합치기
vod_89 = pd.concat([vod_08, vod_09], ignore_index=True)
vod_89
```

```python
# 각 데이터프레임에서 subsr 컬럼의 unique 값 추출
unique_subsr_con_89 = set(con_89['subsr'].unique())
unique_subsr_vod_89 = set(vod_89['subsr'].unique())

# vod_89에만 존재하는 subsr 찾기
subsr_only_in_vod_89 = unique_subsr_vod_89 - unique_subsr_con_89

subsr_only_in_vod_89
```

```python
filtered_rows = vod_89[vod_89['subsr'].isin(subsr_only_in_vod_89)]
filtered_rows
```

```python
filtered_rows.to_csv('../data/전처리data/filtered_rows.csv', index = 0)
```
